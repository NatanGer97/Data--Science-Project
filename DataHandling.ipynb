# Handling And Cleaning Data Set

First of all we noticed that we have columns that we have no use and need for.
Therefore we  delete them.

```python
#we have no use for those columns so we drop them
cleanDataFrame.drop('Index', 1, inplace=True)
cleanDataFrame.drop('ESRB Descriptors', 1, inplace=True)
```

## Missing Data

We saw after examining the data that we have a lot of missing values.
After an examination we found out that columns which have  over then 12000 missing values are columns that we have no use for anyway and do not contribute to our analysis.
As a conclusion , we decided drop them.
In addition, As shown below we have found out that we have features which have more than 50%  of missing values , Therefore we  decided to delete them.
 Moreover, they should be deleted and not to be handled with  because most of them are values which cannot be filled in informedly. for example **"SpecialControllers"** this column will be hard to fill, because it require  to search for each specific game and check whether it has SpecialControllers or not.


```python
labels = cleanDataFrame.isna().sum().index.tolist()
data = cleanDataFrame.isna().sum().values.tolist()
result =  pd.DataFrame({"Attribute":labels,"Amount of nulls":data})
result['% of missing value'] =  (result['Amount of nulls'] /  cleanDataFrame.shape[0] ) * 100
result
```

```python
#Droping all columns with more than 12000 missing values
cleanDataFrame.dropna(axis=1, how='all', thresh=12000, inplace=True)
cleanDataFrame.info()
```

The date column appears in a format that is uncomfortable for us so we transform it to a more readable format 


```python

#Changing the format of dates in release date for more redability
cleanDataFrame['Release Date'] = pd.to_datetime(dataFrame['Release Date']).dt.strftime("%d/%m/%Y") 
cleanDataFrame['Release Date']
```

## Formattin and converting  Columns to Numeric 
Change numeric columns that contain string  to a different format by using regex
And Change columns that  supposed to be numerical but are not, to numeric columns

```python
#Changing strings to "numeric strings"
cleanDataFrame['NumberOfCritics'].replace(to_replace= r'\D+',value='',regex=True,inplace=True)
cleanDataFrame['NumberOfUsers'].replace(to_replace= r'\D+',value='',regex=True,inplace=True)
cleanDataFrame['PositiveUserScore'].replace(to_replace= r'\D',value='',regex=True,inplace=True)
cleanDataFrame['NegativeUserScore'].replace(to_replace= r'\D',value='',regex=True,inplace=True)
```

```python
#Converting object columns to numeric columns
cleanDataFrame["MetaScore"] = pd.to_numeric(cleanDataFrame["MetaScore"],errors='coerce')
cleanDataFrame["NumberOfCritics"] = pd.to_numeric(cleanDataFrame["NumberOfCritics"],errors='coerce')
cleanDataFrame["NumberOfUsers"] = pd.to_numeric(cleanDataFrame["NumberOfUsers"],errors='coerce')
cleanDataFrame["NegativeUserScore"] = pd.to_numeric(cleanDataFrame["NegativeUserScore"],errors='coerce')
cleanDataFrame["MixedUserScore"] = pd.to_numeric(cleanDataFrame["MixedUserScore"],errors='coerce')
cleanDataFrame["PositiveUserScore"] = pd.to_numeric(cleanDataFrame["PositiveUserScore"],errors='coerce')
cleanDataFrame["UserScore"] = pd.to_numeric(cleanDataFrame["UserScore"],errors='coerce')

cleanDataFrame.info()
```

```python
cleanDataFrame.describe()
```

## Filing Missing data
#### Filling missing values by column mean for numeric columns or by a predefine value for categorical columns


```python

columnsToFill = ['NumberOfUsers', 'PositiveUserScore', 'MixedUserScore','NegativeUserScore']
for column in columnsToFill:
       cleanDataFrame[column].fillna(int(cleanDataFrame[column].mean()),inplace=True)
cleanDataFrame['UserScore'].fillna(float(str(cleanDataFrame['UserScore'].mean())[:3]),inplace=True)
cleanDataFrame['Rating'].fillna('Not-Rated',inplace=True)
cleanDataFrame['Developer'].fillna('Not-Specified',inplace=True)
cleanDataFrame['Awards'].fillna('No-Awards',inplace=True)

cleanDataFrame.info()
```

## Splitting  multi-values columns And Creating Indicators Columns for specific attribute

##   For Example:
<p> Awards Column contains in some rows more than one awards. from this rows we descided to split this data to 3 columns that will have only the first 3 awards for each game.<p>

Create  column which indicates if the game have won #1 award

```python
#Adding FirstPlaceAward which indicates if the game have won #1 award
hasWonNumberOneInAnyAward = []
for value in cleanDataFrame['Awards']:
    if "#1" in value:
        hasWonNumberOneInAnyAward.append(1)
    else:
        hasWonNumberOneInAnyAward.append(0)
cleanDataFrame["FirstPlaceAward"] = hasWonNumberOneInAnyAward
```

```python
#The awards column contains a up to 3 awards for each game , we want to split them into 3 diffrent new columns
firstAward = []
secondAward = []
thirdAward = []

for item in cleanDataFrame["Awards"]:
    lst = str(item).split('||')
    try:
        firstAward.append(lst[0])
    except:
        firstAward.append(np.nan)
    try:
        secondAward.append(lst[1])
    except:
        secondAward.append(np.nan)
    try:
        thirdAward.append(lst[2])
    except:
        thirdAward.append(np.nan)

cleanDataFrame["FirstAward"] = firstAward
cleanDataFrame["SecondAward"] = secondAward
cleanDataFrame["ThirdAward"] = thirdAward

cleanDataFrame.drop('Awards', 1, inplace=True)
```

### Release Season indicator

<li>1-Spring</li>
<li>2-Summer</li>
<li>3-Fall</li>
<li>4-Winter</li>

Adding ReleaseSeason column which indicates in which season the game was published


```python
#1-Spring , #2-Summer , #3-Fall , #4-Winter
seasonsDict = {"1":["3","4","5"] , "2":["6","7","8"] , "3":["9","10","11"] , "4":["12","1","2"]}
gameSeasonList = []
for value in cleanDataFrame['Release Date']:
    dateValue = datetime.datetime.strptime(value, "%d/%m/%Y")
    for key,value in seasonsDict.items():
        if str(dateValue.month) in value:
            gameSeasonList.append(key)
cleanDataFrame["ReleaseSeason"] = gameSeasonList
cleanDataFrame["ReleaseSeason"] = pd.to_numeric(cleanDataFrame["ReleaseSeason"],errors='coerce')
```

 ## Separated Genres into Main Genre


The genres column contains multiple genres , we decided to extract the first genre and call it the MainGenre (our assumption was that the first genre is the main genre of the game) , this column will help us to check whether there is an impact of the game genre on reviews amount or rating score.

```python
#Slicing the first genre in Genres and building a new column for it, we only keep the first one because it seems like the most "dominant genre" is first
mainGenre = []
for value in cleanDataFrame["Genres"]:
    mainGenreStr = value.split(",")[0]
    mainGenre.append(mainGenreStr)

cleanDataFrame["MainGenre"] = mainGenre

#Replacing genres which appears less the 200 times in the dataframe to "Other"
for key,value in cleanDataFrame["MainGenre"].value_counts().items():
    if value < 200:
        cleanDataFrame.loc[cleanDataFrame["MainGenre"] == key, "MainGenre"] = "Other"
```

```python
newColumnOrder = ['Title','Release Date','ReleaseSeason','Platform','Developer','Genres','MainGenre','Rating',
'FirstAward','SecondAward','ThirdAward','FirstPlaceAward','MetaScore','NumberOfCritics','PositiveMetaScore'
,'MixedMetaScore','NegativeMetaScore','UserScore','NumberOfUsers','PositiveUserScore','MixedUserScore','NegativeUserScore']

#Changing the order of columns in the dataframe
cleanDataFrame = cleanDataFrame.reindex(columns=newColumnOrder)
cleanDataFrame.head()
```
